## Progress column analysis

#### Only 5 tests have progressed 100

```python
conditions = [
    (df['Progress'] == 100),
    (df['Progress'] >= 50) & (df['Progress'] < 100),
    (df['Progress'] < 50)
]
categories = ['100', '>=50 and <99', '<50']
df['ProgressCategory'] = pd.cut(df['Progress'], bins=[-1, 49, 99, 100], labels=['<50', '>=50 and <99', '100'])

category_counts = df['ProgressCategory'].value_counts().reindex(categories)

plt.figure(figsize=(10, 6))
bars = plt.bar(category_counts.index, category_counts.values, color=['blue', 'orange', 'green'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom')

plt.xlabel('Progress Category')
plt.ylabel('Number of Responses')
plt.title('Number of Responses by Progress Category')
plt.show()
```

## StartDate EndDate RecordedDate Analysis

#### Majority of tests are performed between 2022-12-01 to 2022-12-08
#### Almost all the number of rsponses are generated after 12:00
#### Test were most actively tested on Tuesdays and least on Fridays

```python
df['StartDate'] = pd.to_datetime(df['StartDate'], format='%d-%m-%Y %H:%M')
df['DateStart'] = df['StartDate'].dt.date
df['HourStart'] = df['StartDate'].dt.hour
df['WeekdayStart'] = df['StartDate'].dt.weekday  # Monday = 0, Sunday = 6

daily_counts_start = df['DateStart'].value_counts().sort_index()

df['EndDate'] = pd.to_datetime(df['StartDate'], format='%d-%m-%Y %H:%M')

df['DateEnd'] = df['EndDate'].dt.date
df['HourEnd'] = df['EndDate'].dt.hour
df['WeekdayEnd'] = df['EndDate'].dt.weekday  # Monday = 0, Sunday = 6

daily_counts_end = df['DateEnd'].value_counts().sort_index()

df['DateRecorded'] = df['RecordedDate'].dt.date
df['HourRecorded'] = df['RecordedDate'].dt.hour
df['WeekdayRecorded'] = df['RecordedDate'].dt.weekday  # Monday = 0, Sunday = 6

daily_counts_rec = df['DateRecorded'].value_counts().sort_index()

# # Calculate the number of days since the start date
# df['DaysSinceStart'] = (df['StartDate'] - df['StartDate'].min()).dt.days
# df['DaysSinceStart']
```

```python
numeric_q_columns = [col for col in df.columns if col.startswith('Q') and pd.api.types.is_numeric_dtype(df[col])]
unique_counts_list = []
for col in numeric_q_columns:
    unique_values = df[col].unique()
    counts = df[col].value_counts()
    for value in unique_values:
        unique_counts_list.append({
            'Column': col,
            'Unique_Values': value,
            'Count': counts.get(value, 0)
        })
unique_counts_df = pd.DataFrame(unique_counts_list)
print(unique_counts_df.head(40))
```
